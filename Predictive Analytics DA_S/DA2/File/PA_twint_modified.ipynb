#Pre-Processing

```python
import re
def clean_tweets(text):
  text = re.sub("RT @[\w]*:","",text)
  text = re.sub("@[\w]*","",text)
  text = re.sub("https?://[A-Za-z0-9./]*","",text)
  text = re.sub("\n","",text)
  return text
df['tweet'] = df['tweet'].apply(lambda x: clean_tweets(x))
search_words = "Covid19"
df['Subject'] = search_words
df.head(20)
```

```python
df['tweet']
```

```python
import pandas as pd

# Load the data into a Pandas DataFrame

# Drop unnecessary columns
df = df.drop(['id', 'conversation_id', 'timezone', 'place', 'near', 'geo', 'source',
              'user_rt_id', 'user_rt', 'retweet_id', 'reply_to', 'retweet_date',
              'translate', 'trans_src', 'trans_dest'], axis=1)

# Rename columns
df = df.rename(columns={'created_at': 'datetime', 'user_id': 'user_id_str', 'tweet': 'text'})

# Convert data types
df['datetime'] = pd.to_datetime(df['datetime'])
df['photos'] = df['photos'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Handle missing values
df['mentions'] = df['mentions'].fillna('[]').apply(lambda x: eval(x))
df['urls'] = df['urls'].fillna('[]').apply(lambda x: eval(x))
df['hashtags'] = df['hashtags'].fillna('[]').apply(lambda x: eval(x))
df['cashtags'] = df['cashtags'].fillna('[]').apply(lambda x: eval(x))
df['quote_url'] = df['quote_url'].fillna('')
df['video'] = df['video'].fillna('')
df['thumbnail'] = df['thumbnail'].fillna('')
```

The unnecessary columns are dropped, and the remaining columns are renamed. The data types of the columns are converted to their appropriate types. Missing or null values in some columns are handled by replacing them with empty lists or strings

```python
import pandas as pd
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('stopwords')

# Load the data into a Pandas DataFrame


# Preprocess the text data
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [token for token in tokens if token not in stop_words and token not in punctuation]
    return tokens

df['text'] = df['text'].apply(preprocess_text)

# Count the occurrence of each word in the text data
word_counts = {}
for tokens in df['text']:
    for token in tokens:
        if token in word_counts:
            word_counts[token] += 1
        else:
            word_counts[token] = 1

# Create a dictionary that maps each word to its count
bag_of_words = pd.DataFrame.from_dict(word_counts, orient='index', columns=['count'])
bag_of_words.index.name = 'word'

# Sort the bag of words by frequency
bag_of_words = bag_of_words.sort_values('count', ascending=False)

# Show the top 20 words in the bag of words
print(bag_of_words.head(20))
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon (if necessary)
nltk.download('vader_lexicon')

# Create a sentiment analyzer object
analyzer = SentimentIntensityAnalyzer()

from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize

sia = SentimentIntensityAnalyzer()

for index, row in df.iterrows():
    text = ' '.join(row['text'])  # convert the list of words to a string
    tokens = word_tokenize(text.lower())
    score = sia.polarity_scores(' '.join(tokens))
    df.loc[index, 'compound_sentiment_score'] = score['compound']


def get_sentiment_score(text):
    text = ' '.join(text)  # join the list of words into a string
    sentiment = analyzer.polarity_scores(text)
    return sentiment['compound']


# Apply the sentiment analysis to your data
df['compound_sentiment_score'] = df['text'].apply(get_sentiment_score)

```

```python
import matplotlib.pyplot as plt

# Define a function to get the sentiment label
def get_sentiment_label(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Add a new column with the sentiment label
df['sentiment_label'] = df['compound_sentiment_score'].apply(get_sentiment_label)

# Create a bar plot of the sentiment distribution
sentiment_counts = df['sentiment_label'].value_counts()
colors = ['green', 'red', 'gray']
labels = ['Positive', 'Negative', 'Neutral']
plt.bar(sentiment_counts.index, sentiment_counts.values, color=colors)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Analysis')
plt.xticks(sentiment_counts.index, labels)
plt.show()
```
